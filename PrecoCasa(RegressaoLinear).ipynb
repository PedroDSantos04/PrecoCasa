```python
x_casa = casa.iloc[:,5:6].values
x_casa
```

```python
y_casa = casa.iloc[:,2].values
y_casa
```

```python
from sklearn.model_selection import train_test_split
x_casa_treinamento, x_casa_teste, y_casa_treinamento, y_casa_teste = train_test_split(x_casa, y_casa, test_size = 0.3, random_state= 0)
```

```python
x_casa_treinamento.shape, y_casa_treinamento.shape
```

```python
x_casa_teste.shape, y_casa_teste.shape
```

```python
from sklearn.linear_model import LinearRegression

regre = LinearRegression()
regre.fit(x_casa_treinamento, y_casa_treinamento)
```

```python
previsoes = regre.predict(x_casa_treinamento)
previsoes
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
grafico = px.scatter(x = x_casa_treinamento.ravel(), y = previsoes)
grafico
```

```python
grafico1 = px.scatter(x = x_casa_treinamento.ravel(), y = y_casa_treinamento)

grafico3 = go.Figure(data=grafico1.data)
grafico3
```

```python
grafico1 = px.scatter(x = x_casa_treinamento.ravel(), y = y_casa_treinamento)
grafico2 = px.line(x=x_casa_treinamento.ravel(), y=previsoes)
grafico2.data[0].line.color = 'purple'
grafico3 = go.Figure(data=grafico1.data + grafico2.data)
grafico3
```

```python
previsoes_teste = regre.predict(x_casa_teste)
```

```python
abs (y_casa_teste - previsoes_teste).mean()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_casa_teste, previsoes_teste)
```

```python
mean_squared_error(y_casa_teste, previsoes_teste)
```

```python
grafico4 = px.scatter(x = x_casa_teste.ravel(), y = y_casa_teste)
grafico5 = px.line(x=x_casa_teste.ravel(), y=previsoes_teste)
grafico5.data[0].line.color = 'purple'
grafico6 = go.Figure(data=grafico4.data + grafico5.data)
grafico6
```
